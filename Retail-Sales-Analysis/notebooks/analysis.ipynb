## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from GitHub and many other sources. To find out more about importing data, and how Colab can be used for data science, see the links below under <a href="#working-with-data">Working with data</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

step2: DATA VISUALIZATION + INSIGHTS

```
#which category gives highest sales??
#sales by category (bar chart)
import matplotlib.pyplot as plt
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```
##profit by category ??? which category gives most profits??
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("profit by category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

Some regions perform better than others

```
##YEAR WISE SALES TREND
df.groupby(df['Order Date'].dt.year)['Sales'].sum().plot()
plt.title("Sales Trend Over Years")
plt.show()
```

```
###TOP 10 CUSTOMERS
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("TOP 10 CUSTOMERS")
plt.xlabel("customer names")
plt.ylabel("sales")
plt.show()
```
